```python
from bs4 import BeautifulSoup
import requests
import time 
```

```python
from selenium import webdriver
import pandas as pd
```

```python
url=("https://www.myntra.com/shoes?plaEnabled=false&rf=Price%3A6612.0_13075.0_6612.0%20TO%2013075.0")
```

```python
driver = webdriver.Chrome('G:\\annaconda\\ankit\\chromedriver.exe')
```

```python

driver.get(url)
```

```python
soup=BeautifulSoup(driver.page_source,"html.parser")
```

```python
Brand=[]
Discription=[]
Price=[]
for d in soup.findAll('li', attrs={'class':'product-base'}):
        p=d.find('h3', attrs={'class':'product-brand'})
        Brand.append(p.text)
        pr=d.find('h4', attrs={'class':'product-product'})
        Discription.append(pr.text)
        sr=d.find('span',attrs={'class':'product-discountedPrice'})
        Price.append(sr)
       
```

```python
soup.find('span',attrs={'class':'product-discountedPrice'}).text
```

```python
data=list(zip(Brand,Discription,Price))
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=["Brand","Discription","Price"])
```
