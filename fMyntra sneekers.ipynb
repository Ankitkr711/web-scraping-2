```python
from bs4 import BeautifulSoup
import requests

amazon_text=("https://www.flipkart.com/search?q=sneakers&sid=osp%2Ccil%2Ce1f&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=RECENT&suggestionId=sneakers%7CMen%27s+Casual+Shoes&requestId=af74ec86-f28c-4e16-9b63-830b9a7fd892&as-searchtext=sn&page=")
```

```python
Brand=[]
Description=[]
Price=[]
Discount=[]
for page in range(1,10):
    page=1
    req=requests.get(amazon_text+str(page))
    soup=BeautifulSoup(req.text,"lxml")
    for d in soup.findAll('div', attrs={'class':'_1xHGtK _373qXS'}):
        p=d.find('div', attrs={'class':'_2WkVRV'})
        Brand.append(p.text)
        pr=d.find('a', attrs={'class':'IRpwTa'})
        Description.append(pr.text)
        ra=d.find('div', attrs={'class':'_30jeq3'})
        Price.append(ra.text)
        sr=d.find('div', attrs={'class':'_3Ay6Sb'})
        Discount.append(sr)
                      
```

```python
data=list(zip(Brand,Description,Price,Discount))
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=["product","Discription","Price","Discount"])
```
